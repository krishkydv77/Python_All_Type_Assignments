Palindrome

```python
s=input("Enter String : ")
a=0
index=0
last=len(s)-1
while(index<len(s)//2):
  if(s[index] != s[last]):
    a=1
    break
  index +=1
  last -=1
if a==0:
  print("palindrome")
else:
  print("not palindrome")
```

Armstrong number & Narcissistic number

```python
n=int(input("Enter number : "))
temp=n
temp2=n
count=0
sum=0
while(temp!=0):
  x=temp%10
  count +=1
  temp = temp//10
print("power=",count)
while(temp2 !=0):
  x=temp2%10
  sum+= x**count
  temp2 = temp2//10
if sum==n:
  print("armstrong")
else:
  print("not aramstrong")
```

Nested loop:-


1.    A loop inside another loop.
2.   The inner loop runs campletely for each one stap of the outer loop
     Outer Loop = Runs first
     Inner Loop = run again and again for every outer loop stap.
    
  

*   Nested loop increase time complexity
*   use carefully for large data

*   Comminly used in pattrens, table, matrix.
*   A nested lop is a loop inside another loop.
    -The inner loop executes fully for each iteration of the outer loop.
