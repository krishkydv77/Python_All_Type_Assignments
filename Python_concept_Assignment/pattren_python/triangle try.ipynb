```python
for row in range(6):
  for col in range(row):
    print("*",end=" ")
  print()
```

```python
for row in range(6):
  for col in range(6,row,-1):
    print("*",end=" ")
  print()
```

```python
for row in range(1,6):
  for col in range(1,row+1):
    if row==5 or col ==1 or row==col  :
      print("*",end=" ")
    else:
      print(" ",end=" ")
  print()
```

```python
for row in range(1,6):
  for col in range(6,row,-1):
    if row==1 or col==6 or row==col-1  :
      print("*",end=" ")
    else:
      print(" ",end=" ")
  print()
```

```python
for row in range(1,6):
  for i in range(6,row,-1):
    print(" ",end=" ")
  for col in range(1,row+1):
    print("*",end=" ")
  print()
```

```python
for row in range(1,6):
  for i in range(6,row,-1):
    print(" ",end=" ")
  for col in range(1,row+1):
    if col==1 or row==5 or row==col :
      print("*",end=" ")
    else:
      print(" ",end=" ")
  print()
```

```python
for row in range(1,6):
  for i in range(1,row):
    print(" ",end=" ")
  for col in range(6,row,-1):
    print("*",end=" ")
  print()
```
