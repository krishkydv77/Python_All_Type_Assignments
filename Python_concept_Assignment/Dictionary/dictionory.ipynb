Write a Python program that generates all substrings of a given string and stores the count of substrings starting at each index in a dictionary.

```python
var="abcd"
mydic= {}

for i in range(0,len(var)):
  stri= " "
  for j in range(i,len(var)):
    stri +=var[j]
    if stri not in mydic:
      mydic[i]=1
    else:
      mydic[i]+=1
    print(stri)

```
