Write a Python program to find all triplets in a list whose sum is equal to 30 using the two‑pointer technique.

```python
mylist = [2,4,5,7,9,13,16]
for i in range(len(mylist)):
  start=i+1
  end=len(mylist)-1
  while start<end:
    sum = mylist[i]+mylist[start]+mylist[end]
    if sum==30:
      print(mylist[i],mylist[start],mylist[end])
      start +=1
      end -=1
    elif sum<30:
      start +=1
    else:
      end -=1

```

Write a Python program to generate a nested list (list of lists) where each inner list contains sequential numbers starting from 5, with lengths increasing in each iteration.

```python
#[[5],[6,7],[8,9,10]]
x=5
newlist = []
for i in range(1,4):
  templist=[]     #recreating the templist again and again
  for j in range(1,i+1):
    templist.append(x)   # adding number s,  6,7 into the temp list
    x+=1
  newlist.append(templist)   #append the temp list to original list
  print(" ")
print(newlist)

```

Write a Python program to generate a nested list (list of lists) where each inner list contains numbers starting from 1 up to the current iteration count.

Write a Python program to generate a triangular pattern of numbers where the first and last positions are numbers, the diagonal is numbers, and the remaining positions are filled with *. Also store the pattern in a nested list.
1

1 2

1 * 3

1 * * 4

1 2 3 4 5

```python
mylist=[]
for i in range(1,6):
  templist=[]
  for j in range(1,i+1):
    if i==j or j==1 or i==5:
      print(j,end=" ")
      templist.append(j)
    else:
      print("*",end=" ")
      templist.append('*')
  mylist.append(templist)
  print()
print(mylist)
```
